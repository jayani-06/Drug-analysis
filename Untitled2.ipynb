```python
import pandas as pd
import numpy as np
import joblib

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, accuracy_score
from xgboost import XGBClassifier
```

```python
train_df = pd.read_csv("train.csv")
valid_df = pd.read_csv("valid.csv")
test_df  = pd.read_csv("test.csv")

print(train_df.shape, valid_df.shape, test_df.shape)
train_df.head()
```

```python
TARGET_COLUMN = "label"

FEATURE_COLUMNS = train_df.select_dtypes(include=[np.number]).columns.tolist()

if TARGET_COLUMN in FEATURE_COLUMNS:
    FEATURE_COLUMNS.remove(TARGET_COLUMN)

print("Number of ADMET features:", len(FEATURE_COLUMNS))
FEATURE_COLUMNS[:10]
```

```python
print(train_df.columns.tolist())
```

```python
TARGET_COLUMN = "Y"

# Use ONLY numeric columns except the target
FEATURE_COLUMNS = train_df.select_dtypes(include=[float, int]).columns.tolist()

if TARGET_COLUMN in FEATURE_COLUMNS:
    FEATURE_COLUMNS.remove(TARGET_COLUMN)

print("Features used:", FEATURE_COLUMNS)
```

```python
# Build feature matrices
X_train = train_df[FEATURE_COLUMNS].values
X_valid = valid_df[FEATURE_COLUMNS].values
X_test  = test_df[FEATURE_COLUMNS].values

# Build target vectors
y_train = train_df[TARGET_COLUMN].values
y_valid = valid_df[TARGET_COLUMN].values
y_test  = test_df[TARGET_COLUMN].values

# Sanity checks
print("Target column:", TARGET_COLUMN)
print("Unique y_train values:", set(y_train))
print("Train X shape:", X_train.shape, "Train y shape:", y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
X_test_scaled  = scaler.transform(X_test)

print("Scaled shapes:", X_train_scaled.shape, X_valid_scaled.shape, X_test_scaled.shape)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=100,
    max_depth=4,
    learning_rate=0.1,
    subsample=0.9,
    colsample_bytree=0.9,
    objective="binary:logistic",
    eval_metric="logloss",
    random_state=42
)

model.fit(X_train_scaled, y_train)

print("✅ Model trained successfully")
print("Model expects features:", model.n_features_in_)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report

# Predict probabilities and classes
y_valid_prob = model.predict_proba(X_valid_scaled)[:, 1]
y_valid_pred = (y_valid_prob >= 0.5).astype(int)

y_test_prob = model.predict_proba(X_test_scaled)[:, 1]
y_test_pred = (y_test_prob >= 0.5).astype(int)

# Metrics
print("Validation Accuracy:", accuracy_score(y_valid, y_valid_pred))
print("Validation ROC-AUC:", roc_auc_score(y_valid, y_valid_prob))

print("\nTest Accuracy:", accuracy_score(y_test, y_test_pred))
print("Test ROC-AUC:", roc_auc_score(y_test, y_test_prob))

print("\nClassification Report (Test):")
print(classification_report(y_test, y_test_pred))
```

```python
import joblib
import json

# Save trained model and scaler
joblib.dump(model, "admet_model.joblib")
joblib.dump(scaler, "admet_scaler.joblib")

print("✅ Model and scaler saved")

# Save metadata (very important for deployment)
metadata = {
    "task": "ADMET prediction",
    "target_column": "Y",
    "features": FEATURE_COLUMNS,
    "model_type": "XGBoostClassifier",
    "threshold": 0.5
}

with open("admet_metadata.json", "w") as f:
    json.dump(metadata, f, indent=2)

print("✅ Metadata saved")
```
